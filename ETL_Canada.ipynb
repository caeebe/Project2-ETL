```python
import pandas as pd
import numpy as np
```

```python
canadafile = "CanadaCSV/canadalimited.csv"
```

```python
canada_df = pd.read_csv(canadafile, encoding="utf-8")
```

```python
canada_df.head()
```

```python
list(canada_df.columns.values)
```

```python
canada_df.drop(columns=['C_SEV', 'C_VEHS', 'C_CONF', 'C_RCFG', 'C_RALN', 'C_TRAF', 'V_ID', 'V_YEAR', 'P_ID', 'P_SEX', 'P_AGE', 'P_PSN', 'P_USER'], inplace=True)
```

```python
canada_df['V_TYPE'] = canada_df['V_TYPE'].replace({'01': 1, '05': 0, '06': 0, '07': 0, '08': 0, '09': 0, '10': 0, '11': 0, '14': 0, '16': 0, '17': 0, '18': 0, '19': 0, '20': 0, '21': 0, '22': 0, '23': 0, 'NN': 0, 'QQ': 0, 'UU': 0, 'XX': 0})
```

```python
canada_df['P_ISEV'] = canada_df['P_ISEV'].replace({'3': 1, '2': 0, '1': 0, 'N': 0, 'U': 0, 'X': 0})
```

```python
canada_df['P_SAFE'] = canada_df['P_SAFE'].replace({'01': 2, '02': 0, '09': 0, '10': 0, '11': 0, '12': 0, '13': 0, 'NN': 1, 'QQ': 1, 'UU': 1, 'XX': 1}) 
```

```python
canada_df['P_SAFE'].unique()
```

```python
canada_df.head()
```

```python
group = canada_df.groupby(['C_YEAR', 'C_MNTH', 'C_WDAY', 'C_HOUR', 'C_WTHR', 'C_RSUR'])
group.sum().reset_index().head()
```

```python
df = group['P_ISEV'].sum().reset_index()
df.head()
```

```python
df2 = group['P_SAFE'].max().reset_index()
df2.head()
```

```python
df3 = group['V_TYPE'].max().reset_index()
df3.head()
```

```python
df_merge1 = df.merge(df2, how='inner')
df_merge1.head()
```

```python
datagrouped = df_merge1.merge(df3, how='inner')
datagrouped.head()
```

```python
datagrouped.rename(columns={'C_YEAR': 'year', 'C_MNTH': 'month', 'C_WDAY': 'day', 'C_HOUR': 'hour', 'C_WTHR': 'weather', 'C_RSUR': 'surface', 'P_ISEV': 'fatality', 'P_SAFE': 'safety_device'}, inplace=True)
```

```python
datagrouped.dtypes
```

```python
datagrouped['surface'] = datagrouped['surface'].replace({'1': 'Normal', '2': 'Wet', '3': 'Snow', '4': 'Snow', '5': 'Icy', '6': 'Sandy', '7': 'Muddy', '8': 'Oil', '9': 'Flooded', 'Q': 'Other', 'U': 'Other', 'X': 'Other'})
```

```python
datagrouped.head()
```

```python
onlyfatalities = datagrouped[datagrouped.fatality != 0]
```

```python
finaldata = onlyfatalities[onlyfatalities.V_TYPE != 0]
```

```python
finaldata.head()
```

```python
finaldata['source'] = 'Canada'
```

```python
renamed = finaldata.reset_index()
```

```python
renamed = renamed.drop(columns=['V_TYPE'])
```

```python
renamed.head()
```

```python
renamed.rename(columns={'index':'accident_id'}, inplace=True)
```

```python
renamed.head()
```

```python
renamed['month'] = renamed['month'].replace({'UU': 0, 'XX': 0})
```

```python
renamed['day'] = renamed['day'].replace({'U': 0, 'X': 0})
```

```python
renamed.head()
```

```python
renamed['weather'].replace({'1': 'Clear', '2': 'Cloudy', '3': 'Rain', '4': 'Snow/Hail', '5': 'Snow/Hail', '6': 'Fog/Smoke', '7': 'Strong Wind', 'Q': 'Other', 'U': 'Other', 'X': 'Other'}, inplace=True)
```

```python
renamed.head()
```

```python
renamed['safety_device'].replace({2: 'Not Used', 1: 'Unknown', 0: 'Used'}, inplace=True)
```

```python
renamed.head()
```
